```r
library(class)
```

```r
f <- read.csv("../datasets/wbc_csv.csv")
head(f)
```

```r
f$diagnosis <- as.factor(f$diagnosis)
```

```r
# Shuffle Data
set.seed(123)
f <- f[sample(nrow(f)), ]
```

```r
n <- function(b)
    {
        (b - min(b)) / (max(b) - min(b))
    }
```

```r
# apply func on 3 to 32 columns
f1 <- f[, 3 : 32]
fnor <- as.data.frame(lapply(f1, n))
```

```r
head(fnor)
```

```r
#data partitioning

train <- fnor[1:469, ]
test <- fnor[470: 569, ]
train_label <- fnor[1: 469, 2]
test_label <- fnor[470: 569, 2]
```

```r
# train
p <- knn(train, test, train_label, k=3)
```

```r
# Confusion Matrix
t <- table(test_label,p)
print(t)
```
