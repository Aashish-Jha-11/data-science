```r
#plot for sepal length for each species

library(ggplot2)
```

```r
d <- iris[which(iris$Species == "virginica"), ]
```

```r
p <- ggplot(iris, aes(x = Petal.Length, y =  Species ))
p
#x, y mapping in `aes`. Aesthetic mapping
```

```r
#plotting what
# geometry layer
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species )) + geom_boxplot()
p
```

```r
# color plot as per species

#fill=Species
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species, fill=Species )) + geom_boxplot()
p
```

```r
# mark outlier as red cross
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species, fill=Species )) + geom_boxplot(outlier.shape = 4, outlier.color = "red", outlier.size = 4)
p
```

```r
# add legend
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species, fill=Species )) + geom_boxplot(outlier.shape = 4, outlier.color = "red", outlier.size = 4) + theme(legend.position = "none")
p
#legend.position = "top", "bottom"
```

```r
# change the lables and title
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species, fill=Species )) + geom_boxplot(outlier.shape = 4, outlier.color = "red", outlier.size = 4) + theme(legend.position = "none") + labs(title = "BoxPlot", x = "sepal length", y = "species")
p
```

```r
# rotate the plot
p <- ggplot(iris, aes(x = Sepal.Length, y =  Species, fill=Species )) + geom_boxplot(outlier.shape = 4, outlier.color = "red", outlier.size = 4) + theme(legend.position = "none") + labs(title = "BoxPlot", x = "sepal length", y = "species") + coord_flip()
p
```

```r

g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total)) 
g
```

```r
# geom_point() is scatetrplot
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total)) + geom_point()
g
```

```r
# change scale to log
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total)) + geom_point() + scale_x_log10() + scale_y_log10()
g
```

```r
# color as per region
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, color = murders$region)) + geom_point() + scale_x_log10() + scale_y_log10()
g
```

```r
#Labels, notice smthg wrong.
# because ggplot is global layer and available to everyone, need scoped

g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, color = murders$region, label = abb)) + geom_point() + scale_x_log10() + scale_y_log10() + labs(title = "ScaterPlot") + geom_text()
g
```

```r
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, label = abb)) + geom_point(aes(color = murders$region)) + scale_x_log10() + scale_y_log10() + labs(title = "ScaterPlot") + geom_text()
g
```

```r
# change text size and other things
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, label = abb)) + geom_point(aes(color = murders$region)) + scale_x_log10() + scale_y_log10() + labs(title = "ScaterPlot") + geom_text(size = 3)
g
```

```r
# point and label side by side
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, label = abb)) + geom_point(aes(color = murders$region)) + scale_x_log10() + scale_y_log10() + labs(title = "ScaterPlot") + geom_text(size = 3, nudge_x = 0.075)
g

#nudge_x: move text on x axis
#nudge_y: move text vertically
```

```r
# Title and label change
g <- ggplot(murders, aes(x = murders$population / 10 ^6, y = murders$total, label = abb)) + geom_point(aes(color = murders$region)) + scale_x_log10() + scale_y_log10()  + geom_text(size = 3, nudge_x = 0.075) + labs(title = "ScaterPlot", x = "population", y = "total")
g
```
