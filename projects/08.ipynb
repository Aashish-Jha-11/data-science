```r
library(ggplot2)
```

```r
f <- read.csv("../datasets/Toy_sales_csv.csv")
head(f)
```

```r
l1 <- lm(f$Unitsales ~ f$Price, f)
s1 <- summary(l1)
s1
```

Residual : Error

Estimate : b0 and b1 are coefficients of regression

Eqution: 114215 - 4913 * (Price)

H0 : B1 = 0

Ha : B1 != 0

if p-val < 0.05
    
    reject H0, i.e B1 = 0


```r
p <- ggplot(f, aes(Price, Unitsales)) + geom_point()
p

p2 <- p + geom_smooth(method = "lm", formula = y ~ x, col = "red", se = T)
p2

#for polynomial regression, y ~ x + x^2 + x^3
```

```r
p1 <- predict(l1)
cat("\n Predicted values for SLR: ",p1)
```

```r
err <- f$Unitsales - p1
cat("\n Error for SLR: ",err)
```
